## Read the "classes" csv file

```python
df1 = pd.read_csv("classes.csv", names=['LabelName', 'Class'], header=None)
df1
```

```python
df1.dtypes
```

```python
df2 = df.merge(df1, on='LabelName', how='left')
```

```python
df2 = df2.drop(['Source', 'Confidence', 'IsOccluded', 'IsTruncated', 'IsGroupOf', 'IsDepiction',
        'IsInside', 'XClick1X', 'XClick2X', 'XClick3X', 'XClick4X', 'XClick1Y',
       'XClick2Y', 'XClick3Y', 'XClick4Y'], axis=1)
```

```python
df2
```

```python
import pandas as pd
df_new = pd.DataFrame(columns=df2.columns)

class_list = ['Person', 'Man', 'Woman','Backpack','Briefcase','Handbag', 'Luggage and bags', 'Suitcase']

for item in class_list:
    df_temp = df2[df2['Class']==item]
    df_new = pd.concat([df_new, df_temp])

df_new
```

```python
df_new['Class'] = df_new['Class'].map(
    {
    'Person': 'person',
    'Man': 'person',
    'Woman': 'person',
    'Backpack': 'backpack',
    'Briefcase': 'handbag',
    'Handbag': 'handbag',
    'Luggage and bags': 'luggage',
    'Suitcase': 'luggage',
     }
)

df_new
```

```python
df_new = df_new.drop(['LabelName'], axis=1)
```

```python
import cv2

for i in target_bags:
    image = cv2.imread(f'./new_data/{i}.jpg')
    h, w, c = image.shape
    df_combined.loc[i, 'Height'] = h
    df_combined.loc[i, 'Width'] = w
```

```python
from pascal_voc_writer import Writer

df_bags_n = df_combined.reset_index()

for image, gp in df_bags_n.groupby('ImageID'):
    writer = Writer(f'{gp.iloc[0,0]}', gp.iloc[0, -1], gp.iloc[0, -2])   # each image ID get the first row
    print(gp.iloc[0, -1])
    for i in range(len(gp)):
        y_min = int(gp.iloc[i,3]*gp.iloc[i, -2])
        x_min = int(gp.iloc[i, 1]*gp.iloc[i, -1])
        y_max = int(gp.iloc[i, 4]*gp.iloc[i, -2])
        x_max = int(gp.iloc[i, 2]*gp.iloc[i, -1])
        writer.addObject(f'{gp.iloc[i, 5]}',  x_min, y_min, x_max, y_max)
    writer.save(f'./bags_pics_annotations/{gp.iloc[0,0]}.xml')
```
